```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('Salaries.csv')
df.head()
```

```python
df.columns
```

- Basic Data Exploration :

```python
print(f"the number of rows in the dataset : {df.shape[0]}")
print(f"the number of features in the dataset : {df.shape[1]}")
```

```python
df.info()
```

- Checking duplicates :

```python
df.duplicated().sum()
```

- Checking missing values :

```python
print("Missing values of each column")
for column_name in df.columns:
    print(f"{column_name} : {df[column_name].isnull().sum()}")
```

- Descriptive Statistics :

```python
mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode().values[0]
min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()
salary_range = max_salary - min_salary
std_dev_salary = df['TotalPay'].std()


print(f"Mean Salary: {mean_salary}")
print(f"Median Salary: {median_salary}")
print(f"Mode Salary: {mode_salary}")
print(f"Minimum Salary: {min_salary}")
print(f"Maximum Salary: {max_salary}")
print(f"Salary Range: {salary_range}")
print(f"Standard Deviation of Salary: {std_dev_salary}")
```

- Data cleaning :

```python
df['BasePay'].isnull().sum()
```

```python
df['BasePay'].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['BasePay'], bins=20, kde=True)
plt.xlabel('BasePay')
plt.ylabel('Frequency')
plt.title(f'BasePay Histogram')
plt.show()
```

```python
df['BasePay'].fillna(df['BasePay'].median(),inplace=True)
```

```python
sns.histplot(df['OvertimePay'], bins=20, kde=True)
plt.xlabel('OvertimePay')
plt.ylabel('Frequency')
plt.title(f'OvertimePay Histogram')
plt.show()
```

```python
df['OvertimePay'].describe()
```

```python
df['OvertimePay'].fillna(df['OvertimePay'].median(),inplace=True)
```

```python
sns.histplot(df['OtherPay'], bins=20, kde=True)
plt.xlabel('OtherPay')
plt.ylabel('Frequency')
plt.title(f'OtherPay Histogram')
plt.show()
```

```python
df['OtherPay'].describe()
```

```python
df['OtherPay'].fillna(df['OtherPay'].median(),inplace=True)
```

```python
sns.histplot(df['Benefits'], bins=20, kde=True)
plt.xlabel('Benefits')
plt.ylabel('Frequency')
plt.title(f'Benefits Histogram')
plt.show()
```

```python
df['Benefits'].describe()
```

```python
df['Benefits'].fillna(df['Benefits'].median(),inplace=True)
```

```python
columns_to_drop = ['Id','Notes', 'Status']
df.drop(columns=columns_to_drop, inplace=True)
```

```python
df.head()
```

- Basic Data Visualization :

```python
sns.histplot(df['TotalPay'], bins=20, kde=True)
plt.xlabel('TotalPay')
plt.ylabel('Frequency')
plt.title(f'TotalPay Histogram')
plt.show()

sns.histplot(df['TotalPayBenefits'], bins=20, kde=True)
plt.xlabel('TotalPayBenefits')
plt.ylabel('Frequency')
plt.title(f'TotalPayBenefits Histogram')
plt.show()
```

```python
# Count the occurrences of each JobTitle
job_title_counts = df['JobTitle'].value_counts()

# Plotting the pie chart
plt.pie(job_title_counts, labels=job_title_counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.title('Proportion of Employees in Different Job Titles')
plt.show()
```

- Grouped Analysis :

```python
columns_of_interest = ['Agency','JobTitle', 'Year']

# Perform grouped analysis for each column using a for loop
for column in columns_of_interest:
    grouped_by_column = df.groupby(column)
    average_salary_by_column = grouped_by_column['TotalPay'].mean(0)
    print(f"Average Salary by {column}:\n", average_salary_by_column)
    print("\n")
```

- Simple Correlation Analysis :

```python
correlation_coefficient = df['TotalPay'].corr(df['Benefits'])

print(f"Correlation between Salary and Benefits : {correlation_coefficient}")
```

```python
# Scatter Plot
plt.scatter(df['Benefits'], df['TotalPay'])
plt.xlabel('Benefits')
plt.ylabel('TotalPay')
plt.title(f'Scatter Plot: TotalPay vs Benefits')
plt.show()
```
